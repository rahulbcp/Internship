Question 1: Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
##Answer 1
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(page.content)
header = []
for title in soup.find_all('span',class_="mw-headline"):
    header.append(title.text)
header_tags = pd.DataFrame({"Titles":header})
header_tags
```

```python
## Answer 3 - a)
cricket_website = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
extract = BeautifulSoup(cricket_website.content)
ODI_team = []
for details in extract.find_all('span',class_="u-hide-phablet")[:10]:
    ODI_team.append(details.text)
ODI_matches_1 = []
for details in extract.find_all('td', class_="rankings-block__banner--matches"):
    ODI_matches_1.append(details.text)
ODI_matches_2 = []
for details in extract.find_all('td', class_="table-body__cell u-center-text")[:18]:
    ODI_matches_2.append(details.text)
Final_ODI_matches = ODI_matches_2[::2]
ODI_matches = ODI_matches_1 + Final_ODI_matches
ODI_Points_1 = []
for details in extract.find_all('td', class_="rankings-block__banner--points"):
    ODI_Points_1.append(details.text)
ODI_Points_2 = []
for details in extract.find_all('td', class_="table-body__cell u-center-text")[:19]:
    ODI_Points_2.append(details.text)
Final_ODI_Points = ODI_Points_2[1::2]
ODI_points = ODI_Points_1 + Final_ODI_Points
ODI_Ratings_1 =[]
for details in extract.find_all('td', class_="rankings-block__banner--rating"):
    ODI_Ratings_1.append(details.text.replace(" ","").replace("\n",""))
ODI_Ratings_2 = []
for details in extract.find_all('td', class_="table-body__cell u-text-right rating")[:9]:
    ODI_Ratings_2.append(details.text)
ODI_rating = ODI_Ratings_1 + ODI_Ratings_2
ODI_mens = pd.DataFrame({"Teams": ODI_team, "Matches": ODI_matches, "Points": ODI_points, "Rating": ODI_rating})
ODI_mens
```

```python
##Answer 3 - b)
batting_website = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
ODI_extract = BeautifulSoup(batting_website.content)
ODI_batsmen_1 = []
for details in ODI_extract.find_all('div', class_="rankings-block__banner--name-large"):
    ODI_batsmen_1.append(details.text)
ODI_batsmen_2 = []
for details in ODI_extract.find_all('td', class_="table-body__cell rankings-table__name name")[:9]:
    ODI_batsmen_2.append(details.text.replace('\n',""))
ODI_batsmen = ODI_batsmen_1 + ODI_batsmen_2
ODI_batsmen_team_1 = []
for details in ODI_extract.find_all('div', class_="rankings-block__banner--nationality"):
    ODI_batsmen_team_1.append(details.text.replace('\n',"").replace(" ", ""))
ODI_batsmen_team_2 = []
for details in ODI_extract.find_all('span', class_="table-body__logo-text")[:9]:
    ODI_batsmen_team_2.append(details.text.replace('\n',"").replace(" ", ""))
ODI_batsmen_team = ODI_batsmen_team_1 + ODI_batsmen_team_2
ODI_batsmen_rating_1 = []
for details in ODI_extract.find_all('div', class_="rankings-block__banner--rating"):
    ODI_batsmen_rating_1.append(details.text)
ODI_batsmen_rating_2 = []
for details in ODI_extract.find_all('td', class_="table-body__cell rating")[:9]:
    ODI_batsmen_rating_2.append(details.text)
ODI_batsmen_rating = ODI_batsmen_rating_1 + ODI_batsmen_rating_2
mens_batsmen = pd.DataFrame({"Batsmen":ODI_batsmen, "Team":ODI_batsmen_team, "Rating":ODI_batsmen_rating})
mens_batsmen
```

```python
##Answer - 4 a)
icc = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
download = BeautifulSoup(icc.content)
women_team = []
for details in download.find_all('span',class_="u-hide-phablet")[:10]:
    women_team.append(details.text)
women_matches_1 = []
for details in download.find_all('td', class_="rankings-block__banner--matches"):
    women_matches_1.append(details.text)
women_matches_2 = []
for details in download.find_all('td', class_="table-body__cell u-center-text")[:18]:
    women_matches_2.append(details.text)
Total_Matches = women_matches_2[::2]
matches = women_matches_1 + Total_Matches
women_points_1 = []
for details in download.find_all('td', class_="rankings-block__banner--points"):
    women_points_1.append(details.text)
women_points_2 = []
for details in download.find_all('td', class_="table-body__cell u-center-text")[:19]:
    women_points_2.append(details.text)
total_points = women_points_2[1::2]
points = women_points_1 + total_points
women_rating_1 = []
for details in download.find_all('td', class_="rankings-block__banner--rating"):
    women_rating_1.append(details.text.replace("\n", "").replace(" ", ""))
women_rating_2 = []
for details in download.find_all('td', class_="table-body__cell u-text-right rating")[:9]:
    women_rating_2.append(details.text)
rating = women_rating_1 + women_rating_2
df = pd.DataFrame({"Team":women_team, "Matches":matches, "Points":points, "Rating":rating})
df    
```

Question 5 - Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-
i) Headline
ii) Time
iii) News Link

```python
#Answer 5
news = requests.get("https://www.cnbc.com/world/?region=world")
fetch = BeautifulSoup(news.content)
headline = []
for details in fetch.find_all('a', class_="LatestNews-headline"):
    headline.append(details.text)
time = []
for details in fetch.find_all('time', class_="LatestNews-timestamp"):
    time.append(details.text)
news_links = []
for details in fetch.find_all('a', href=True)[223:265]:
    news_link = details['href']
    news_links.append(news_link.replace("/pro/", "").replace("/investingclub/", ""))
df = pd.DataFrame({"News Link":news_links})
df_cleaned = df[df['News Link'] != '']
News_link_1 = df_cleaned.values.tolist()
df = pd.DataFrame({"Headline":headline, "Time":time, "News Link":News_link_1})
df
```

Question 6 -Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame-
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL

```python
#Answer 6
articles = requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
articles
AI = BeautifulSoup(articles.content)
AI
paper_title = []
for details in AI.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    paper_title.append(details.text)
len(paper_title)
authors = []
for details in AI.find_all('span', class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(details.text)
authors
published_date = []
for details in AI.find_all('span', class_="sc-1thf9ly-2 dvggWt"):
    published_date.append(details.text)
len(published_date)
paper_urls = []
for details in AI.find_all('a', href=True)[53:77]:
    paper_url = details['href']
    paper_urls.append(paper_url)
len(paper_urls)
df = pd.DataFrame({"Paper Title":paper_title, "Authors":authors, "Published Date":published_date, "Paper URL":paper_urls})
df
```

Question 7 - Write a python program to scrape mentioned details from dineout.co.in and make data frame-
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
#Answer 7
dine = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
dine
scrape = BeautifulSoup(dine.content)
scrape
restaurant_name = []
for details in scrape.find_all("a", class_="restnt-name ellipsis"):
    restaurant_name.append(details.text)
restaurant_name  
cuisine = []
for details in scrape.find_all("span", class_="double-line-ellipsis"):
    cuisine_details = details.text.split('|', 1)[1].strip()
    cuisine.append(cuisine_details)
cuisine    
location = []
for details in scrape.find_all('div', class_="restnt-loc ellipsis"):
    location.append(details.text)
location    
rating = []
for details in scrape.find_all('div', class_="restnt-rating rating-4"):
    rating.append(details.text)
rating    
image_url = []
for details in scrape.find_all('img', class_="no-img"):
    image_url.append(details["data-src"])
image_url   
df = pd.DataFrame({"Restaurant name":restaurant_name, "Cuisine":cuisine, "Location":location,"Ratings":rating, "Image URL":image_url})
df
```
